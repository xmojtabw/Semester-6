```python
from matplotlib import pyplot as plt
```

```python
def EWA(points,beta,v0=20):
    output = []
    previous = v0
    for p in points:
        current = (beta * previous) + ((1-beta)* p)
        output.append(current)
        previous = current
    return output
```

```python
points = [22,24,20,19,21]
ewa9 = EWA(points,0.9)
ewa5 = EWA(points,0.5)
```

```python
# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(points, 'o-', label='Original Points', markersize=8)
plt.plot(ewa9, 's-', label='EWA (=0.9)', markersize=8)
plt.plot(ewa5, 's-', label='EWA (=0.5)', markersize=8)


# Add labels and title
plt.title('Exponentially Weighted Average (EWA) Visualization')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Show the plot
print("EWA with beta 0.9: ", ewa9)
print("EWA with beta 0.5: ", ewa5)

plt.show()
```
