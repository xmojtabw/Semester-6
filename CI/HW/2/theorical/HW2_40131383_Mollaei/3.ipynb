```python
import numpy as np 
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
data  = np.array([[0.5,1.5,2.0],[1.0,2.0,1.0]])
labels = np.array([0.0,1.0,0.0]).reshape(1,3)

landa = 0.01
learning_rate = 0.1


w1 = np.array([[0.2,-0.3],[0.4 , 0.1]])
b1 = np.array([0.1,-0.2]).reshape((2,1))

w2 = np.array([0.3,-0.5]).reshape((2,1))
b2 = np.array([0.05]).reshape((1,1))

```

```python
def forward(data,w1,b1,w2,b2):
    z1 = w1.T @ data + b1 
    a1 = sigmoid(z1)
    print('a1 : ', a1,'\n ---')
    z2 =  w2.T @ a1 + b2  
    a2 = sigmoid(z2) 
    print('a2 : ', a2,'\n ---')
    return a1,a2
    

def cost(predicted , label,landa):
    m = labels.shape[1]
    l2 = (landa/(2*m)) * (np.sum(w1* w1)+np.sum(w2*w2)) 
    loss = - (label *np.log( predicted) + (1-label) * np.log(1-predicted))
    return np.mean(loss) + l2    
```

```python
a1 , predictions = forward(data,w1,b1,w2,b2)
print ("cost is ", cost(predictions,labels,landa))
```

```python
def backward(data,w1,w2,a2,a1,labels): 
    m = labels.shape[1]
    dz2 =  a2 - labels 
    dw2 = ((1/m)* (a1 @ dz2.T)) + ((landa/m)*(w2))
    db2 = np.mean(dz2)
    da1 =   w2 @ dz2  
    dz1 = da1 * a1 *(1- a1)
    dw1 = ((1/m )*  data @ dz1.T) + ((landa/m)*(w1))
    db1  = np.mean(dz1,axis=1).reshape((2,1)) 
    return dw2 , db2 ,dw1,db1  
```

```python
def train(data,w1,w2,b1,b2,epochs,learning_rate,landa):
    for i in range(epochs):
        print(f'epoch {i+1} :')
        a1 , a2 = forward(data,w1,b1,w2,b2)
        print ("cost is ", cost(a2,labels,landa))
        dw2 , db2 , dw1 , db1 = backward(data,w1,w2,a2,a1,labels)
        w1 -= learning_rate * dw1
        w2 -= learning_rate * dw2
        b1 -= learning_rate * db1
        b2 -= learning_rate * db2 
        print("updated weights w1\n",w1 , "\nb1:\n",b1,"\nw2:\n ", w2,'\nb2:\n',b2)

```

```python
train(data,w1,w2,b1,b2,3,learning_rate,landa)
```
