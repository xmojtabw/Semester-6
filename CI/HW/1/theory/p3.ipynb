```python
def f(x):
    return x**4 + 4*x**3 - 688*x**2 + 2672*x - 2640

def df(x):
    return 4*x**3 + 12*x**2 - 1376*x + 2672
```

```python
x = 5.5

# running gradian decent
steps = 5 # number of steps
learning_rate= 0.0026
points = [x]
for i in range(steps+1):
    print(i, '&', f'{x:.3f}', '&', int(f(x)), '&', int(df(x)), '\\\\')
    x = x - learning_rate * df(x)
    points.append(x)
```

```python
from matplotlib import pyplot as plt 

x = range(-40, 40)
y = [f(i) for i in x]
dy = [df(i) for i in x]
plt.plot(x, y, label='f(x)')
plt.plot(x, dy, label='df(x)')
plt.plot(points, [f(i) for i in points], 'ro', label='steps')
```

```python
import math as m
def f2(x):
    return 1/10000 * (x-10)*(x-3)*(x-1)*(x+1)*(x+3)*(x+10) * (300*m.exp( -1*abs(x) )*((1+m.cos(x))/2)+1)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Vectorize the function to handle numpy arrays
f2_vectorized = np.vectorize(f2)

# Generate x values from -40 to 40
x = np.linspace(-40, 40, 400)

# Compute y values using the vectorized function f2
y = f2_vectorized(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)')
plt.xlim(-15, 15)
plt.ylim(-30, 50)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
