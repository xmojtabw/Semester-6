```python
def f(x):
    return x**4 + 4*x**3 - 688*x**2 + 2672*x - 2640

def df(x):
    return 4*x**3 + 12*x**2 - 1376*x + 2672
```

```python
x = 5.5

# running gradian decent
steps = 5 # number of steps
learning_rate= 0.0026
points = [x]
for i in range(steps+1):
    print(i, '&', f'{x:.3f}', '&', int(f(x)), '&', int(df(x)), '\\\\')
    x = x - learning_rate * df(x)
    points.append(x)
```

```python
from matplotlib import pyplot as plt 

x = range(-40, 40)
y = [f(i) for i in x]
dy = [df(i) for i in x]
plt.plot(x, y, label='f(x)')
plt.plot(x, dy, label='df(x)')
plt.plot(points, [f(i) for i in points], 'ro', label='steps')
```
